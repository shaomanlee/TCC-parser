```python
#import required modules
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time
import re
import sys
import pandas as pd
```

```python
#separate cells for try and error
#fix JY110
#cf 80, 90, 213, 229, 419, 585, 627, 633, 645, & 110
#removed 129, 211, 347 刪除事後補(註)
#679句首加回兩全形空格
#744亂碼 更改為粧品

#print(req.encoding) #当前编码，默认为 ISO-8859-1
#print(req.apparent_encoding) #当前网页的内容的实际编码


# the period chosen: x - (y-1) 號解釋  解釋文
for n in range(1, 776): 
    req = requests.get('https://www.judicial.gov.tw/constitutionalcourt/p03_01.asp?expno='+str(n))
    req.encoding ='Big5'
    src = req.text
    soup = BeautifulSoup(src, 'lxml')
    rows = soup.select('tr tr tr tr')
    holdingdata = rows[3].select('div.expreson_content')
    # Keep only the text in each of those cells
    holding = [cell.text for cell in holdingdata]
    saveholding = pd.Series(holding)
    saveholding.to_csv("jyh_%s.txt" %n,encoding="utf-8", index=False)
    print (n)
```
